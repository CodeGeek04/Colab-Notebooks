```python
from collections import Counter

def frequency(column):
  _, unique = pd.factorize(df[column])
  class_freq = Counter(_.flatten())
  tot = 0
  sorted_freq = sorted(class_freq.items(), key=lambda x: x[1], reverse=True)
  ind = 1
  for i, freq in sorted_freq:
    print(ind, ". ", unique[i], ": ", freq, sep = "", end="\n__________________________________________________________________\n")
    tot += freq
    ind += 1
  print("TOTAL =", tot)


def correlation(column1, column2):
  df2 = df.groupby([column1, column2])
  print(df2.size())
```

```python
frequency('INJ_BODY_PART')
```

```python
correlation("CLASSIFICATION", "INJ_BODY_PART")
```
